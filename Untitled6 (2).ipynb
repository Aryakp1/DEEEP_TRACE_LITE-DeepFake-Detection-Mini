```python
import os

def print_dir_structure(start_path, max_depth=3):
    for root, dirs, files in os.walk(start_path):
        # Calculate depth
        depth = root[len(start_path):].count(os.sep)
        if depth > max_depth:
            continue
        indent = ' ' * 2 * depth
        print(f"{indent}{os.path.basename(root)}/")
        for f in sorted(files):
            print(f"{indent}  {f}")

# Change this to your actual path
base_path = '/content/drive/MyDrive/DeepTraceLite'
print_dir_structure(base_path)
```

```python
import os
import cv2
from collections import defaultdict
from tqdm import tqdm

def extract_clips_from_frames(src_folder, dest_folder, label):
    os.makedirs(dest_folder, exist_ok=True)

    # Group all frames per video ID
    videos = defaultdict(list)
    for filename in sorted(os.listdir(src_folder)):
        if filename.endswith(".jpg"):
            parts = filename.replace('.jpg', '').split('_f')
            if len(parts) == 2:
                video_id = parts[0]
                frame_idx = int(parts[1])
                videos[video_id].append((frame_idx, filename))

    clip_counter = 0
    for video_id, frames in tqdm(videos.items(), desc=f"Processing {label}"):
        # Sort frames by index
        sorted_frames = sorted(frames, key=lambda x: x[0])
        num_frames = len(sorted_frames)

        # Create 5 clips: overlapping if needed
        for clip_num in range(5):
            if clip_num + 5 <= num_frames:
                clip_frames = sorted_frames[clip_num:clip_num + 5]
            else:
                # Pad with last frame if not enough
                clip_frames = sorted_frames[-5:]

            clip_dir = os.path.join(dest_folder, f"{label}_{clip_counter}")
            os.makedirs(clip_dir, exist_ok=True)

            for i, (_, fname) in enumerate(clip_frames):
                src_path = os.path.join(src_folder, fname)
                dst_path = os.path.join(clip_dir, f"f{i}.jpg")
                img = cv2.imread(src_path)
                if img is not None:
                    cv2.imwrite(dst_path, img)
            clip_counter += 1

# Paths
real_frame_path = "/content/drive/MyDrive/DeepTraceLite/frames/real"
fake_frame_path = "/content/drive/MyDrive/DeepTraceLite/frames/fake"
real_clip_path = "/content/drive/MyDrive/DeepTraceLite/clip_dataset/real"
fake_clip_path = "/content/drive/MyDrive/DeepTraceLite/clip_dataset/fake"

# Run for both real and fake
extract_clips_from_frames(real_frame_path, real_clip_path, "real")
extract_clips_from_frames(fake_frame_path, fake_clip_path, "fake")


```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.models as models
from torchvision.models import MobileNet_V2_Weights
from torch.utils.data import Dataset, DataLoader, Subset
import torchvision.transforms as transforms
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import train_test_split
import os
from PIL import Image
from tqdm import tqdm
import numpy as np
import random

# Set seed
torch.manual_seed(42)
random.seed(42)
np.random.seed(42)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# --------------------------
# Dataset class
# --------------------------
class DeepTraceLiteDataset(Dataset):
    def __init__(self, root_dir, label, transform=None):
        self.root_dir = root_dir
        self.transform = transform
        self.label = label
        self.clips = [os.path.join(root_dir, clip) for clip in sorted(os.listdir(root_dir))]

    def __len__(self):
        return len(self.clips)

    def __getitem__(self, idx):
        clip_path = self.clips[idx]
        frames = []
        for i in range(5):  # 5 frames per clip
            frame_path = os.path.join(clip_path, f'f{i}.jpg')
            image = Image.open(frame_path).convert('RGB')
            if self.transform:
                image = self.transform(image)
            frames.append(image)
        clip_tensor = torch.stack(frames)  # Shape: (T=5, C, H, W)
        label = torch.tensor(float(self.label), dtype=torch.float32)
        return clip_tensor, label

# --------------------------
# Focal Loss
# --------------------------
class FocalLoss(nn.Module):
    def __init__(self, alpha=1, gamma=2, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.bce = nn.BCEWithLogitsLoss(reduction='none')
        self.reduction = reduction

    def forward(self, inputs, targets):
        BCE_loss = self.bce(inputs, targets)
        pt = torch.exp(-BCE_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * BCE_loss
        return focal_loss.mean() if self.reduction == 'mean' else focal_loss.sum()

# --------------------------
# Model
# --------------------------
class DeepTraceLiteModel(nn.Module):
    def __init__(self, hidden_size=128):
        super(DeepTraceLiteModel, self).__init__()
        weights = MobileNet_V2_Weights.DEFAULT
        base = models.mobilenet_v2(weights=weights)
        for param in base.parameters():
            param.requires_grad = False
        base.eval()
        self.feature_extractor = base.features
        self.avgpool = nn.AdaptiveAvgPool2d((1, 1))
        self.gru = nn.GRU(input_size=1280, hidden_size=hidden_size, batch_first=True)
        self.classifier = nn.Linear(hidden_size, 1)

    def forward(self, x):
        B, T, C, H, W = x.shape
        x = x.view(B * T, C, H, W)
        with torch.no_grad():
            feats = self.feature_extractor(x)
        feats = self.avgpool(feats).view(B, T, -1)
        out, _ = self.gru(feats)
        out = out[:, -1, :]
        return self.classifier(out).squeeze(1)

# --------------------------
# Transforms
# --------------------------
transform = transforms.Compose([
    transforms.Resize((160, 160)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5]*3, std=[0.5]*3)
])

# --------------------------
# Prepare Dataset + Split
# --------------------------
real_dataset = DeepTraceLiteDataset("/content/drive/MyDrive/DeepTraceLite/clip_dataset/real", 0, transform)
fake_dataset = DeepTraceLiteDataset("/content/drive/MyDrive/DeepTraceLite/clip_dataset/fake", 1, transform)

# Combine
full_dataset = real_dataset.clips + fake_dataset.clips
full_labels = [0] * len(real_dataset) + [1] * len(fake_dataset)

# Stratified split
train_idx, val_idx = train_test_split(
    list(range(len(full_dataset))),
    test_size=0.2,
    stratify=full_labels,
    random_state=42
)

# Custom dataset wrapper to access both clips and labels
class CombinedDeepTraceDataset(Dataset):
    def __init__(self, clip_paths, labels, transform):
        self.clip_paths = clip_paths
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.clip_paths)

    def __getitem__(self, idx):
        clip_path = self.clip_paths[idx]
        frames = []
        for i in range(5):
            frame_path = os.path.join(clip_path, f'f{i}.jpg')
            image = Image.open(frame_path).convert('RGB')
            if self.transform:
                image = self.transform(image)
            frames.append(image)
        clip_tensor = torch.stack(frames)
        label = torch.tensor(float(self.labels[idx]), dtype=torch.float32)
        return clip_tensor, label

train_dataset = CombinedDeepTraceDataset([full_dataset[i] for i in train_idx], [full_labels[i] for i in train_idx], transform)
val_dataset   = CombinedDeepTraceDataset([full_dataset[i] for i in val_idx],   [full_labels[i] for i in val_idx],   transform)

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)
val_loader   = DataLoader(val_dataset, batch_size=8)

# --------------------------
# Train Loop
# --------------------------
model = DeepTraceLiteModel().to(device)
criterion = FocalLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
num_epochs = 15

for epoch in range(num_epochs):
    model.train()
    total_loss = 0
    all_preds, all_labels = [], []

    for clips, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}"):
        clips, labels = clips.to(device), labels.to(device)
        outputs = model(clips)
        loss = criterion(outputs, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
        preds = (torch.sigmoid(outputs) > 0.5).cpu().numpy()
        all_preds.extend(preds)
        all_labels.extend(labels.cpu().numpy())

    train_acc = accuracy_score(all_labels, all_preds)
    train_auc = roc_auc_score(all_labels, all_preds)
    print(f"\nEpoch {epoch+1}/{num_epochs}")
    print(f"Train Loss: {total_loss:.4f} | Acc: {train_acc:.4f} | AUC: {train_auc:.4f}")

    # --------------------------
    # Validation
    # --------------------------
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for clips, labels in val_loader:
            clips, labels = clips.to(device), labels.to(device)
            outputs = model(clips)
            preds = (torch.sigmoid(outputs) > 0.5).cpu().numpy()
            all_preds.extend(preds)
            all_labels.extend(labels.cpu().numpy())

    val_acc = accuracy_score(all_labels, all_preds)
    val_auc = roc_auc_score(all_labels, all_preds)
    print(f"Validation Acc: {val_acc:.4f} | AUC: {val_auc:.4f}")


```

```python
import cv2
import numpy as np

def extract_clips(video_path, clip_length=10, num_clips=5):
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    clips = []
    step = max(total_frames // num_clips, 1)

    for i in range(num_clips):
        frames = []
        start_frame = i * step
        cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)
        for _ in range(clip_length):
            ret, frame = cap.read()
            if not ret:
                break
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            frames.append(frame)
        if len(frames) == clip_length:
            clips.append(np.array(frames))
    cap.release()
    return clips
```

```python
import torch
import torch.nn.functional as F
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# ---- 1. Global Variables to Store Hooks ----
gradients = None
activations = None

def save_gradients_hook(module, grad_input, grad_output):
    global gradients
    gradients = grad_output[0]

def save_activations_hook(module, input, output):
    global activations
    activations = output

# ✅ Register to the last conv layer of MobileNetV2
# model.feature_extractor is Sequential
target_layer = model.feature_extractor[-1]
target_layer.register_forward_hook(save_activations_hook)
target_layer.register_full_backward_hook(save_gradients_hook)

# ---- 2. Grad-CAM Generator ----
def generate_gradcam(input_clip):
    global gradients, activations
    gradients, activations = None, None

    input_clip = input_clip.to(device)  # Shape: (1, T, C, H, W)
    input_clip.requires_grad = True

    # Forward pass
    model.eval()
    output = model(input_clip)  # Shape: (1, 1)

    # Backward pass
    model.zero_grad()
    output.backward()

    if gradients is None or activations is None:
        print("❌ Gradients or activations not captured. Check hooks.")
        return np.zeros((input_clip.shape[-2], input_clip.shape[-1]))

    pooled_gradients = torch.mean(gradients, dim=[0, 2, 3])  # (C,)
    activations_ = activations.detach().squeeze(0)  # (C, H, W)

    for i in range(activations_.shape[0]):
        activations_[i, :, :] *= pooled_gradients[i]

    heatmap = torch.mean(activations_, dim=0).cpu().numpy()
    heatmap = np.maximum(heatmap, 0)

    if np.max(heatmap) != 0:
        heatmap /= np.max(heatmap)

    return heatmap

# ---- 3. Overlay Heatmap ----
def overlay_heatmap_on_frame(frame, heatmap, alpha=0.4):
    heatmap = cv2.resize(heatmap, (frame.shape[1], frame.shape[0]))
    heatmap_color = cv2.applyColorMap(np.uint8(255 * heatmap), cv2.COLORMAP_JET)
    overlayed = cv2.addWeighted(frame, 1 - alpha, heatmap_color, alpha, 0)
    return overlayed

# ---- 4. Run Grad-CAM on Video ----
def gradcam_on_video(video_path, output_path="gradcam_output.mp4", clip_len=5):
    model.eval()
    clips = extract_clips(video_path, clip_length=clip_len, num_clips=1)
    frame_list = []

    for clip in clips:
        # Transform all frames in the clip
        transformed_clip = [transform(Image.fromarray(f)).unsqueeze(0) for f in clip]  # List of (1, C, H, W)
        clip_tensor = torch.cat(transformed_clip, dim=0).unsqueeze(0).to(device)  # (1, T, C, H, W)

        # ✅ Generate Grad-CAM on full clip
        heatmap = generate_gradcam(clip_tensor)

        # Middle frame for overlay
        middle_idx = clip_len // 2
        middle_frame = clip[middle_idx]
        overlay = overlay_heatmap_on_frame(middle_frame, heatmap)
        frame_list.append(overlay)

    if not frame_list:
        print("❌ No frames processed.")
        return

    # Save to video
    height, width, _ = frame_list[0].shape
    out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), 5, (width, height))
    for frame in frame_list:
        out.write(cv2.cvtColor(frame, cv2.COLOR_RGB2BGR))
    out.release()

    print(f"✅ Grad-CAM video saved to: {output_path}")

# ---- 5. Run ----
gradcam_on_video("/content/deepfake_dataset/train_sample_videos/aagfhgtpmv.mp4")




```

```python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import cv2

# --- Grad-CAM Class ---
class GradCAM:
    def __init__(self, model, target_layer):
        self.model = model
        self.model.eval()
        self.target_layer = target_layer

        self.activations = None
        self.gradients = None
        self._register_hooks()

    def _register_hooks(self):
        def forward_hook(module, input, output):
            print("✅ Forward hook triggered")
            self.activations = output

        def backward_hook(module, grad_input, grad_output):
            print("✅ Backward hook triggered")
            self.gradients = grad_output[0]

        self.target_layer.register_forward_hook(forward_hook)
        self.target_layer.register_full_backward_hook(backward_hook)

    def generate(self, input_clip, target_class=None):
        self.model.eval()
        input_clip = input_clip.unsqueeze(0).to(device)  # (1, T, C, H, W)

        # Patch model forward (for feature extraction before GRU)
        def patched_forward(x):
            B, T, C, H, W = x.shape
            x = x.view(B * T, C, H, W)
            feats = self.model.feature_extractor(x)                   # (B*T, features, H, W)
            feats = self.model.avgpool(feats).view(B, T, -1)          # (B, T, features)
            out, _ = self.model.gru(feats)
            out = out[:, -1, :]                                       # last time step
            return self.model.classifier(out).squeeze(1)

        original_forward = self.model.forward
        self.model.forward = patched_forward

        input_clip.requires_grad_(True)
        output = self.model(input_clip)
        self.model.forward = original_forward

        if target_class is None:
            target = torch.ones_like(output)  # assume fake
        else:
            target = torch.tensor([target_class], dtype=torch.float32).to(device)

        self.model.zero_grad()
        output.backward(gradient=target)

               # Compute CAM
        grads = self.gradients                    # (B*T, C, H, W)
        activs = self.activations                 # (B*T, C, H, W)

        weights = torch.mean(grads, dim=(2, 3), keepdim=True)  # Global average pooling -> (B*T, C, 1, 1)
        cam = torch.sum(weights * activs, dim=1)               # Weighted sum -> (B*T, H, W)
        cam = F.relu(cam)

        # Normalize each CAM map individually
        cam_min = cam.amin(dim=(1, 2), keepdim=True)
        cam_max = cam.amax(dim=(1, 2), keepdim=True)
        cam = (cam - cam_min) / (cam_max - cam_min + 1e-8)

        # Reshape back to (T, H, W)
        B, T = 1, input_clip.shape[1]
        cam = cam.view(B, T, cam.shape[1], cam.shape[2])[0]  # (T, H, W)

        return cam
```

```python
def apply_colormap_on_image(img, activation_map, colormap_name='JET'):
    # Resize activation_map to match image size
    activation_map = cv2.resize(activation_map, (img.shape[1], img.shape[0]))

    # Apply heatmap
    heatmap = cv2.applyColorMap(np.uint8(255 * activation_map), getattr(cv2, f'COLORMAP_{colormap_name}'))
    heatmap = np.float32(heatmap) / 255

    # Overlay heatmap on image
    cam = heatmap + np.float32(img)
    cam = cam / np.max(cam)
    return np.uint8(255 * cam)

```

```python
import numpy as np

def smooth_predictions(preds, window_size=3):
    return np.convolve(preds, np.ones(window_size)/window_size, mode='same')
```

```python
import torch
import numpy as np

def predict_video_from_clips(model, clip_list, device='cuda', apply_smoothing=False, window_size=3):
    model.eval()
    probs = []

    with torch.no_grad():
        for clip in clip_list:
            clip = clip.unsqueeze(0).to(device)  # [1, T, C, H, W]
            output = model(clip)
            prob = torch.sigmoid(output).item()
            probs.append(prob)

    probs = np.array(probs)
    if apply_smoothing:
        probs = smooth_predictions(probs, window_size=window_size)

    final_prob = np.mean(probs)  # Mean after (optional) smoothing
    return final_prob, probs  # Return full list of clip probs for visualization
```

```python
from torchvision import transforms
import cv2
import torch

def extract_clips_from_video(video_path, num_clips=5, frames_per_clip=10, transform=None):
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    frame_idxs = np.linspace(0, total_frames - 1, num=num_clips * frames_per_clip, dtype=int)

    clips = []
    frames = []

    i = 0
    for idx in frame_idxs:
        cap.set(cv2.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        if not ret:
            continue
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        frame = cv2.resize(frame, (160, 160))
        if transform:
            frame = transform(frame)
        else:
            frame = torch.tensor(frame / 255.0).permute(2, 0, 1).float()

        frames.append(frame)
        i += 1
        if i % frames_per_clip == 0:
            clip = torch.stack(frames[-frames_per_clip:])  # shape: [T, C, H, W]
            clips.append(clip)

    cap.release()
    return clips
```

```python
import torch
import torch.nn as nn
from torchvision import models

class DeepTraceLiteModel(nn.Module):
    def __init__(self, hidden_size=64, num_layers=1):
        super(DeepTraceLiteModel, self).__init__()
        mobilenet = models.mobilenet_v2(pretrained=False)
        self.features = mobilenet.features  # Feature extractor
        self.pool = nn.AdaptiveAvgPool2d((1, 1))

        self.gru = nn.GRU(input_size=1280, hidden_size=hidden_size, num_layers=num_layers, batch_first=True)
        self.classifier = nn.Linear(hidden_size, 1)

    def forward(self, x):  # x: [B, T, C, H, W]
        B, T, C, H, W = x.size()
        x = x.view(B * T, C, H, W)
        with torch.no_grad():
            features = self.features(x)
            features = self.pool(features).view(B, T, -1)  # [B, T, 1280]

        out, _ = self.gru(features)  # [B, T, hidden]
        last_output = out[:, -1, :]  # use last timestep
        logits = self.classifier(last_output)  # [B, 1]
        return logits

def load_model(model_path, device='cpu'):
    model = DeepTraceLiteModel()
    model.load_state_dict(torch.load(model_path, map_location=torch.device(device)))
    model.to(device)
    model.eval()
    return model

```
